An end to end Scikit-Learn workflow

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
heart_disease = pd.read_csv("/content/drive/MyDrive/Datasets/43 - heart-disease.csv")
heart_disease
```

```python
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
# from sklearn.linear import LinearRegression
clf = RandomForestClassifier()
clf2 = HistGradientBoostingClassifier()
```

splitting data

```python
from sklearn.model_selection import train_test_split
X = heart_disease.drop("target", axis=1) # means columns for axis 1, we will drop it
X

```

```python
y = heart_disease["target"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
clf.fit(X_train, y_train)
```

```python
clf2.fit(X_train, y_train)
# clf2.score(X_test, y_test
```

```python
# clf.get_params()
```

```python
X_train
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_pred2 = clf2.predict(X_test)
y_pred2
```

```python
clf2.score(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf2.score(X_test, y_test)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_preds))
print(classification_report(y_test, y_preds))
```
